```python
zoodata_1 = zoodata.drop('animal name',axis=1)
zoodata_1.head()
```

```python
x = zoodata_1.drop('type',axis = 1)
y = zoodata_1[['type']]
```

```python
x.shape,y.shape
```

```python
scale = StandardScaler()
std = scale.fit_transform(x)
x_scale = pd.DataFrame(std,columns=x.columns)
x_scale.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scale,y,test_size=0.20,random_state=40)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
model = KNeighborsClassifier()
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
y_train_pred
```

```python
train_accuracy = accuracy_score(y_train,y_train_pred)
train_accuracy
```

```python
y_test_pred = model.predict(x_test)
y_test_pred
```

```python
test_accuracy = accuracy_score(y_test,y_test_pred)
test_accuracy
```

```python
n_neighbors = np.array(range(1,50))
param_grid  = dict(n_neighbors=n_neighbors)
```

```python
model=KNeighborsClassifier()
grid =GridSearchCV(estimator=model,param_grid=param_grid)
grid.fit(x_scale,y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=12)
cv_scores = []
for i in range (1,50,2):
    knn_model = KNeighborsClassifier(n_neighbors=i)
    cross_val_scores =cross_val_score(estimator=knn_model,X=x_scale,y=y,cv=kfold)
    print(i,'th Iteration:\n',cross_val_scores.mean().round(4))
    cv_scores.append(cross_val_scores.mean().round(4))
```

```python
max(cv_scores)
```

```python
plt.plot(range(1,50,2),cv_scores)
plt.xlabel('Neighbors')
plt.ylabel('CV Scores')
plt.title('Neighbors VS CV scores')
plt.show()
```

```python
plt.scatter(range(1,50,2),cv_scores)
plt.xlabel('Neighbors')
plt.ylabel('CV Scores')
plt.title('Neighbors VS CV scores')
plt.show()
```
