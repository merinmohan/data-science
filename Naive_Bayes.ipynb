```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
salarydata_train = pd.read_csv('SalaryData_train.csv')
salarydata_train.head()
```

```python
salarydata_train.shape
```

```python
salarydata_train.info()
```

```python
salarydata_train.describe(include='all')
```

```python
corr_matrix = salarydata_train.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
le = LabelEncoder()
salarydata_train['workclass'] = le.fit_transform(salarydata_train['workclass'])
salarydata_train['education'] = le.fit_transform(salarydata_train['education'])
salarydata_train['maritalstatus'] = le.fit_transform(salarydata_train['maritalstatus'])
salarydata_train['occupation'] = le.fit_transform(salarydata_train['occupation'])
salarydata_train['relationship'] = le.fit_transform(salarydata_train['relationship'])
salarydata_train['race'] = le.fit_transform(salarydata_train['race'])
salarydata_train['sex'] = le.fit_transform(salarydata_train['sex'])
salarydata_train['native'] = le.fit_transform(salarydata_train['native'])
salarydata_train['Salary'] = le.fit_transform(salarydata_train['Salary'])
```

```python
salarydata_train.dtypes
```

```python
corr_matrix_1 = salarydata_train.corr()
corr_matrix_1
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix_1,annot= True)
plt.show()
```

```python
salarydata_train.drop('workclass',inplace=True,axis=1)
salarydata_train.drop('education',inplace=True,axis=1)
salarydata_train.drop('occupation',inplace=True,axis=1)
salarydata_train.drop('race',inplace=True,axis=1)
salarydata_train.drop('native',inplace=True,axis=1)
```

```python
salarydata_train.head()
```

```python
x_train = salarydata_train.iloc[:,:-1]
y_train = salarydata_train.iloc[:,8:]
```

```python
scale = StandardScaler()
std = scale.fit_transform(x_train)
x_std_train = pd.DataFrame(std,columns=x_train.columns)
```

```python
x_std_train.head()
```

```python
nb = GaussianNB()
nb.fit(x_std_train,y_train)
```

```python
y_train_pred = nb.predict(x_std_train)
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
train_accuracy
```

### Training Accuracy is 0.7925

```python
confusion_matrix(y_train,y_train_pred)
```

```python
print('Classification_Report for train data :\n ',classification_report(y_train,y_train_pred))
```
