```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
moviesdata = pd.read_csv('my_movies.csv')
moviesdata
```

```python
moviesdata.shape
```

```python
moviesdata.drop(moviesdata.iloc[:,0:5],axis=1,inplace=True)
```

```python
moviesdata.head()
```

### 1.Association rule with 10% support and 80% confidence

```python
freq_itemsets = apriori(moviesdata,min_support=0.1,use_colnames=True)
freq_itemsets['length'] = freq_itemsets['itemsets'].apply(lambda x: len(x))
freq_itemsets
```

```python
rule_1 = association_rules(freq_itemsets,metric='lift',min_threshold=0.8)
rule_1
```

```python
rule_1.sort_values('lift',ascending=False)
```

```python
sns.scatterplot(x=rule_1['support'],y=rule_1['confidence'],data=rule_1)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
